# readind the file

```python
import pandas as pd

file_path = r"C:\Users\AFAK\Desktop\OnlineRetail.csv"
df = pd.read_csv(file_path, encoding='latin1')  # أو 'ISO-8859-1'

df

```

# formating the Date

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

# checking the values

```python
df=df[df['UnitPrice']>=0]
df=df[df['Quantity']>=0]
```

```python
df.describe()
```

# dropping duplicate values

```python
df.drop_duplicates()
```

```python
df.dropna()
```

```python
group_by_year_season=df.groupby(['year','season'])['revenu'].sum()
group_by_year_season=group_by_year_season.reset_index()
group_by_year_season
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# الحصول على السنوات الفريدة
years = df['year'].unique()

# إنشاء الشكل
plt.figure(figsize=(12, 6))

# رسم خط لكل سنة
for year in years:
    yearly_data = group_by_year_season[group_by_year_season['year'] == year]
    plt.plot(yearly_data['season'], yearly_data['revenu'], marker='*', label=f'{year}')

# إعدادات الرسم
plt.legend()
plt.title('Beziehung zwischen Menge und jahreszeit per jaher')
plt.xlabel('season')
plt.ylabel('revenu')
plt.grid(True)
plt.show()
```
